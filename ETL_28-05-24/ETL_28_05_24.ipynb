```python
# NEEV SHAH - 22BTRCL107
import numpy as np
import pandas as pd

data = pd.read_csv("/content/data.csv")
# Step-3:
# a
cleaned = data.drop_duplicates()
# b
cleaned['Calories'].fillna(cleaned['Calories'].mean(), inplace=True)
# c
cleaned = cleaned[(cleaned['Pulse'].between(50, 200)) & (cleaned['Maxpulse'].between(60, 220))]

# Step-4
cleaned.rename(columns={
    'Duration': 'Exercise_Time',
    'Pulse': 'Pulse_Rate',
    'Maxpulse': 'Maximum_Pulse_Rate',
    'Calories': 'Calories_Burned'
}, inplace=True)

# Step 5
output_file_path = 'cleaned_data.csv'
cleaned.to_csv(output_file_path)

print(f"Cleaned dataset saved to: {output_file_path}")
cleaned.head()

```
